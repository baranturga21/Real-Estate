```python
df_2 = pd.read_excel(r"ilçe.xlsx")
df_2.drop("Unnamed: 0", axis = 1, inplace = True)
df = df_2.copy()
```

```python
df.head()
```

```python
df=df.dropna(subset=['TotalAppraisedValue'])
```

```python
df_3 = df.copy()
df_4 = df.copy()
```

```python
from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
```

```python
df_4["StreetNameAndWay"] = le.fit_transform(df_3.StreetNameAndWay)
```

```python
le.classes_
```

```python
X = df_4.drop(["SalePrice"], axis = 1)
y = df_4["SalePrice"]
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV,cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd
```

```python
from sklearn import model_selection
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 144)
```

```python
import xgboost as xgb
from xgboost import XGBRegressor
```

```python
xgb1 = XGBRegressor(colsample_bytree = 0.6, learning_rate = 0.1, max_depth = 2, n_estimators = 200)
```

```python
model_xgb = xgb1.fit(X_test, y_test)
```

```python
model_xgb.predict(X_test)[15:20]
```

```python
model_xgb.score(X_train, y_train)
```
